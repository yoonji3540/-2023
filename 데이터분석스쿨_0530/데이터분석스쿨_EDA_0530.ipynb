### 이상치 확인하기

- 이상치(outlier) : 사분위에서 특정(+-IQR*1.5) 범위를 초과하는 경우


```python
import matplotlib.pyplot as plt
```

```python
# 이상치 확인 -> boxplot

plt.boxplot(out_df)
plt.show()
```

```python
out_df['국어'] = out_df['국어'].replace(810, 81)
out_df
```

```python
plt.boxplot(out_df)
plt.show()
```

```python
# 학생들의 국어(수학, 영어) 성적 평균

out_df.mean()

plt.bar(out_df.mean().index, out_df.mean())
# plt.bar(out_df.columns, out_df.mean())
```

```python
# 과목별 최댓값

out_df.max()
```

```python
# 과목별 최솟값

out_df.min()
```

```python
# 순서대로 -> 사분위값 -> 상위 몇 %
# q1(25%), q2(50%), q3(75%)

q1 = out_df.quantile(0.25)
q2 = out_df.quantile(0.5)
q3 = out_df.quantile(0.75)
IQR = q3 - q1 # -> 50%에 해당하는 데이터

print(q1)
print(q2)
print(q3)
print(IQR)
```

```python
# 일반적인 데이터 허용 범위:(-1.5 * IQR, 1.5 * IQR)
# 이 범위를 벗어나면 이상치(outlier)
```

```python
limit = IQR * 1.5

#  - limit < 데이터 < + limit
# for문을 사용하여 outlier인지 판단하기

'''
if 값 > + limit or 값 < - limit:
  return outlier

else:
  return data
'''
```

```python
# groupby: 기준 데이터 -> (처리할)수치 데이터
# 월 별 / 기온, 풍속 ...
```

```python
weather_df.index # DatatimeIndex
```

```python
weather_date = pd.Series(weather_df.index)
weather_date
```

```python
weather_date.dt.month
```

```python
weather_df['월'] = weather_date.dt.month.values
```

```python
weather_df
```

```python
# groupby
month_low_temp = weather_df.groupby(weather_df['월'])['최저기온(°C)'].mean()
```

```python
# 최고기온
month_high_temp = weather_df.groupby(weather_df['월'] )['최고기온(°C)'].mean()

# 일강수량 
month_daily_vol = weather_df.groupby(weather_df['월'])['일강수량(mm)'].mean()

# 최대 순간 풍속
month_wind_speed = weather_df.groupby(weather_df['월'])['최대 순간 풍속(m/s)'].mean()

# 평균 풍속
month_wind_mean = weather_df.groupby(weather_df['월'])['평균 풍속(m/s)'].mean()

# 최대 순간 풍속 풍향
month_wind_dic = weather_df.groupby(weather_df['월'])['최대 순간 풍속 풍향(hhmi)'].mean()
```

```python
# x_month = weather_df['월']
# x_month['']

x_month = weather_df['월'].unique()
```

```python
# 그룹바이 월별데이터 -> 월별 데이터들 -> 그래프
plt.figure(figsize=(10, 10))

# 1
plt.subplot(2, 3, 1)
plt.title('월별 최저기온')
plt.ylabel('기온')
plt.xlabel('월')
plt.plot(x_month, month_low_temp)

# 2
plt.subplot(2, 3, 2)
plt.title('월별 최고기온')
plt.ylabel('기온')
plt.xlabel('월')
plt.plot(x_month, month_high_temp)

# 3
plt.subplot(2, 3, 3)
plt.title('월별 일강수량')
plt.ylabel('강수량')
plt.xlabel('월')
plt.plot(x_month, month_daily_vol)

# 4
plt.subplot(2, 3, 4)
plt.title('최대 순간 풍속')
plt.ylabel('속도')
plt.xlabel('월')
plt.plot(x_month, month_wind_speed)

# 5
plt.subplot(2, 3, 5)
plt.title('평균 풍속')
plt.ylabel('속도')
plt.xlabel('월')
plt.plot(x_month, month_wind_mean)

# 5
plt.subplot(2, 3, 6)
plt.title('최대순간 풍속 풍향')
plt.ylabel('도')
plt.xlabel('월')
plt.plot(x_month, month_wind_dic)



plt.show()
```

```python
# 기온 그래프를 하나의 그래프로 표시
plt.figure(figsize=(5, 5))


plt.plot(x_month, month_low_temp, label='최저기온')
plt.plot(x_month, month_high_temp, ':' , label='최고기온')
plt.legend(loc='upper right')
# 색상: (b blue), (r red), (y yellow), (w white)
# 선: (-- dash line), (-. dash dot), (: dot)
# 범례 위치: upper left, uppter center, center(left, right), lower(left, right)

plt.show()
```

```python
# 일강수량 그래프를 바 그래프로 나타내기

plt.figure(figsize=(5, 5))
plt.title('월별 강수량')
plt.ylabel('mm')
plt.xlabel('월')
plt.bar(x_month, month_daily_vol)

plt.show()
```

```python
# 일강수량 그래프를 바 그래프로 나타내기(가로 그래프로)
plt.figure(figsize=(5, 5))
plt.title('월별 강수량')
plt.xlabel('mm')
plt.ylabel('월')
plt.barh(x_month, month_daily_vol)

plt.show()
```
