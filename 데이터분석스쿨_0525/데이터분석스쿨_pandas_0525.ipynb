## XML 데이터

- eXtensible Markup Language
- HTML과 유사 -> 화면 구성
- XML -> 데이터 전송


```
<person>
  <name>date</name>
</person>
```
(꺽새: 태그 = 요소)
(안에 값: 속성)

### XML 파일 불러오기

````python
xml = pd.read_xml('raw_data.xml')
xml
````

````python
# <element></element> 요소
# <element 속성='값'></element>
# <element>'값'</element>
````
