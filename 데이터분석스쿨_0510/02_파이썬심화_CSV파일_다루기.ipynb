```python
f = open('list.pickle' , 'wb')
test = [1,2,3,4,5]
pickle.dump(test, f)
f.close()
```

```python
f = open('list.pickle', 'rb')
test_pickle = pickle.load(f)
print(test_pickle)
```

```python
# pickle은 클래스도 저장가능하다.

class Multiply(object):
  def __init__(self, multi):
    self.multi = 5
  def multiply(self, number):
    return number * self.multi

multi_test = Multiply(5)
multi_test.multiply(10)
```

```python
f = open('multiply_object.pickle', 'wb')
pickle.dump(multi_test, f)
f.close()
```

```python
f = open('multiply_object.pickle', 'rb')
multi_test = pickle.load(f)
```

```python
multi_test.multiply(10)
```

```python
# 많은 데이터를 처리할 수 있는 함수 만들기

def switch(listname):
  for i in listname: #이중 리스트안에 있는 리스트 1개를 가져오고
    #['Jongrogu', 151,767, 11,093, 27,394]
    for j in i: #리스트에 있는 요소를 받아와서
      try: # 여기 해당되는건 바꾸고 (, 문자형을 숫자형으로)
        i[i.index(j)] = float(re.sub(',','',j))
      except: # 해당되지 않으면 바꾸지 말아라(숫자가 없는 문자형은 넘어가라)
        pass
  return listname
```

```python
# 가장 더웠던 날을 출력해보자!!
import csv
f = open('CSV_example2_data.csv')
data = csv.reader(f)

max_temp = 0

header = next(data)
for row in data:
  row[4] = float(row[4]) # 문자형 데이터를 숫자형으로 바꾸고
  
  if max_temp < row[4]: #만약 지금까지 최고 기온값보다 현재 값의 최고기온 값이 더 크다면
    max_temp = row[4] #최고기온
    max_data = row[0] #최고기온날짜
f.close()
```

# Matplotlib에 대해 알아보기

```python
# 01. Matplotlib 기본 사용방법

matplotlib.pyplot 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음입니다.

matplotlib.pyplot 모듈의 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있습니다.

예를 들어, 그래프 영역을 만들고, 몇 개의 선을 표현하고, 레이블로 꾸미는 등의 일을 할 수 있습니다.
```

```python
# y축 지정하기
# as 새로운 이름 지정할때 사용합니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4])
plt.show() # print()

pyplot.plot() 함수에 하나의 숫자 리스트를 입력함으로써 아래와 같은 그래프가 그려집니다.

plot() 함수는 리스트의 값들이 y 값들이라고 가정하고, x 값 [0, 1, 2, 3]을 자동으로 만들어냅니다.

matplotlib.pyplot 모듈의 show() 함수는 그래프를 화면에 나타나도록 합니다.
```

```python
# x축 y축 지정하기
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-')
plt.axis([1, 5, 0, 20])
plt.show()
```

```python
x, y 값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열 (Format string)을 세번째 인자에 입력할 수 있습니다.

포맷 문자열 ‘ro’는 빨간색 (‘red’)의 원형 (‘o’) 마커를 의미합니다.

또한, 예를 들어 ‘b-‘는 파란색 (‘blue’)의 실선 (‘-‘)을 의미합니다.

matplotlib.pyplot 모듈의 axis() 함수를 이용해서 축의 범위 [xmin, xmax, ymin, ymax]를 지정했습니다.

결과는 아래와 같습니다.
```

```python
이 예제는 연도별로 변화하는 값을 갖는 데이터를 막대 그래프로 나타냅니다.

[0,1,2] 주어진 범위와 간격에 따라 균일한 값을 갖는 리스트를 만들었습니다.

years는 X축에 표시될 연도이고, values는 막대 그래프의 y 값 입니다.

먼저 plt.bar() 함수에 x 값 [0, 1, 2]와 y 값 [100, 400, 900]를 입력해주고,

xticks()에 x와 years를 입력해주면, X축의 눈금 레이블에 ‘2018’, ‘2019’, ‘2020’이 순서대로 표시됩니다.

결과는 아래와 같습니다.
```

```python
import os
os.chdir('/content/')

import csv

f = open('/content/CSV_example2_data.csv', 'r') #csv를 리드 모드로 불러옴
data = csv.reader(f)
next(data)

# 4,5월 데이터 리스트
result_04 = [] 
result_05 = []

for row in data: 
  # 월만 뽑아오기
  month = row[0].split('-')[1] # 0 연도 1 월 2 일
  if month == '04':
    result_04.append(float(row[-1]))  
  if month == '05':
    result_05.append(float(row[-1]))  
f.close()

import matplotlib.pyplot as plt

#plt.figure(figsize = (10, 2)) #그래프의 가로길이를 10인치 세로를 2인치로 설정
plt.hist(result_04,color = 'b', bins = 10) 
plt.hist(result_05,color = 'r', bins = 10) 
plt.show()
```

```python
# 서울 특별시 0세부터 100세까지의 연령에 따른 선 그래프 그리기

import csv
f = open('/content/dataset_example3_new.csv')
data = csv.reader(f)

seoul = []
for row in data:
  if '경기도' == row[0]: #서울특별시를 포함한 것을 찾기
    for i in row[3:14]: # 0~100세까지 산출
      i = i.replace(",","") # 대체
      seoul.append(int(i))
f.close()
```

```python
import csv
f = open('/content/dataset_example3_new.csv')
data = csv.reader(f)

female = [] #여성 인구정보를 담을 리스트
male = [] #남성 인구정보를 담을 리스트

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[16:27]:#1에서 100까지
      i = i.replace(",", "")
      male.append(int(i)) #남성데이터는 15~27
    for i in row[29:]:
      i = i.replace(",", "")
      female.append(int(i)) #여성데이터는 30~번까지

import matplotlib.pyplot as plt

plt.bar(range(11), male)
plt.bar(range(11), female)
plt.legend()
plt.show()
f.close()
```

```python
import csv
f = open('/content/dataset_example3_new.csv')
data = csv.reader(f)

female = [] #여성 인구정보를 담을 리스트
male = [] #남성 인구정보를 담을 리스트

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[16:27]:#1에서 100까지
      i = i.replace(",", "")
      male.append(-int(i)) #남성데이터는 15~27
    for i in row[29:]:
      i = i.replace(",", "")
      female.append(int(i)) #여성데이터는 30~번까지

import matplotlib.pyplot as plt

plt.barh(range(11), male)
plt.barh(range(11), female)
plt.legend()
plt.show()
f.close()
```

```python
# 마지막 연습문제 
# 리스트 3개를 만드셔서 -> 연도 / 서울 / 수도권
# 리스트로 값을 저장하는 쿼리를 작성시면 됩니다.

year = []
s = []
g = []

with open('example_4_all.csv') as f:
  reader = csv.reader(f)
  for row in reader:
    if row[0] != "구분":
      year.append(row[0])
      s.append(float(row[1]))
      g.append(float(row[2]))
```

```python
print(year)
print(s)
print(g)
```

```python
# 그래프 연습문제 마지막

# year가 x축의 이름
# s g를 2개의 그래프를 그리는 연습문제입니다.

import matplotlib.pyplot as plt

plt.plot(range(14), s, label = 's')
plt.plot(range(14), g, label = 'g')

plt.xticks(range(14), year, rotation = 45)

plt.show()
```

```python
import matplotlib.pyplot as plt
# x축에 년도, y축에 인구수가 있는 선 그래프 만들기
plt.plot(range(14), s, color='green', marker='o', linestyle='solid', label='s')
plt.plot(range(14), g, color='red', marker='o', linestyle='solid', label='g')

plt.xticks(range(14), year, rotation=45)

plt.title("Population Change")
plt.ylabel("Percentage")

plt.legend()
plt.show()
```
