##### 해변 데이터 중 검색량의 이상치 판별

혼잡도로 표현

```python
# q1, q2, q3가 있을 때 (q3 -q1) * 1.5 를 초과하면 이상치 -> 검색량이 너무 많다 -> 혼잡도가 높다

plt.boxplot(nature['검색건수'])
plt.show()
```

```python
# 상위 경계값
q1 = 6737
q3 = 53095
upper_IQR = q3 + (q3 - q1) * 1.5
upper_IQR
```

```python
import numpy as np
```

```python
# 상위 경계값을 초과하는 값들만 지워준다
nature['혼잡도'] = np.where(nature['검색건수'] > upper_IQR, '혼잡', '보통')
nature
```
