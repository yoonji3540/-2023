### Groupby

- 기준: 범주형 데이터(남과 여, 지역별, 혈액형, 국적, 학년, 반, mbti ...)
- 파악할 데이터: 수치형 데이터

- 데이터를 그룹별로 분할

1. split: 원본 데이터에서 분류하고자하는 데이터를 나눠준다
2. apply: 분류된 데이터(그룸별)에 함수 적용
3. combine: 함수 적용된 데이터를 다시 합쳐줌

### 데이터프레임 병합

```python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }, index=['a', 'b', 'c', 'd']
)

data2 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]},
    index=['e', 'f', 'g', 'h']
    )
```

```python
add_data = pd.concat([data, data2]) # 두 개의 데이터를 합쳐줌
# add_data = pd.concat([data, data2], ignore_index=True) ignore_index=True: 인덱스가 겹칠경우 인덱스 재배열
# concat: 기본적으로 행으로 붙는다(axis=0) 
# 행으로 붙음 -> 행이 추가됨 /  열으로 붙음 -> 열이 추가됨
```

```python
add_data
```

```python
df1 = pd.DataFrame([['A0', 'A1', 'A2', 'A3'],
                     ['B0', 'B1', 'B2', 'B3'],
                     ['C0', 'C1', 'C2', 'C3'],
                     ['D0', 'D1', 'D2', 'D3']], 
                   columns=list('ABCD'))
 
df2 = pd.DataFrame([['A4', 'A5', 'A6', 'A7'],
                     ['B4', 'B5', 'B6', 'B7'],
                     ['C4', 'C5', 'C6', 'C7'],
                     ['D4', 'D5', 'D6', 'D7']], 
                   columns=['A', 'B', 'C', 'D'], index=[4, 5, 6, 7])

df3 = pd.DataFrame([['A8', 'A9', 'A10', 'A11'],
                     ['B8', 'B9', 'B10', 'B11'],
                     ['C8', 'C9', 'C10', 'C11'],
                     ['D8', 'D9', 'D10', 'D11']], 
                   columns=list('ABCD'), 
                   index=[8, 9, 10, 11])
```

```python
print(df1)
print(df2)
print(df3)
```

```python
pd.concat([df1, df2, df3]) # default: 행으로 붙는다
```

```python
pd.concat([df1, df2, df3], axis=1) # axis=1 열로 붙는다
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                  'F': ['F2', 'F3', 'F6', 'F7']}, 
                   index=[2, 3, 6, 7])
```

Join option

```python
pd.concat([df1, df4], axis=1, join='outer') # concat의 default는 join='outer'
# outer: 합집합
```

```python
pd.concat([df1, df4], axis=1, join='inner')
# inner: 교집합
```

```python
pd.concat([df1, df4], axis=1).reindex(df1.index) # left join (df1을 기준으로 합친다)
```

```python
pd.concat([df1, df4], axis=1).reindex(df4.index) # right join (df4를 기준으로 합친다)
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                  'A': ['A0', 'A4', 'A2', 'A3'],
            'B': ['B0', 'B4', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                  'C': ['C0', 'C1', 'C2', 'C3'],
                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
print(left)
print(right) # key 값 중에 겹치는 값이 있음
```

```python
mi = pd.merge(left, right) # 기본적으로 열으로 붙음
# default: inner
```

```python
mi
```

```python
mo = pd.merge(left, right, how='outer')
```

```python
print(mo)
```

```python
m_key = pd.merge(left, right, on='key') # mi = pd.merge(left, right)와 결과가 같음
# on으로 원하는 기준을 설정할 수 있음
```

```python
m_key 
```

```python
m_l = pd.merge(left, right, on='key', how='left')
m_l
```

```python
m_r = pd.merge(left, right, on='key', how='right')
m_r
```
